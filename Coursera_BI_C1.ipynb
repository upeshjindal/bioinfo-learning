```python
def count_polymers(text, pattern):
    
    count = 0
    matches = re.finditer(pattern=pattern, string=text)
    for match in matches:
        count += 1
        # print(match.start())
        
    return count
```

```python
count = count_polymers(text="CGTGACAGTGTATGGGCATCTTT", pattern="TGT")
print (count)
```

```python
def hamming_distance(str1, str2):
    
    dist = 0
    for i,j in zip(str1, str2):
        if i != j:
            dist += 1
        
    return dist
```

```python
hamming = hamming_distance(
    "CTTGAAGTGGACCTCTAGTTCCTCTACAAAGAACAGGTTGACCTGTCGCGAAG",
    "ATGCCTTACCTAGATGCAATGACGGACGTATTCCTTTTGCCTCAACGGCTCCT",
)
print(f"{hamming=}")
```

```python
def calculate_skew(gnome: str):
    skew = 0
    max_skew = -1
    max_i = 0
    skew_list = {}
    for i in range(len(gnome)):
        
        if gnome[i] == "G":
            skew += 1
        elif gnome[i] == "C":
            skew -= 1
            
        skew_list[i] = skew
            
        if skew > max_skew:
            max_skew = skew
            max_i = i
            
    return skew, max_skew, max_i, skew_list
```

```python
skew,max_skeq,max_i,skew_dist = calculate_skew("CATTCCAGTACTTCATGATGGCGTGAAGA")
print(max_i)
```

```python
def frequency(gnome, nueclotide):
    freq = 0
    for i in range(len(gnome)):
        if gnome[i] == nueclotide:
            freq += 1
            
    return freq / len(gnome) * 100 - 25
```

```python
def freq(gnome, nucleotide):
    c_freq = []
    f = 0
    for i in range(len(gnome)):
        if gnome[i] == nucleotide:
            if i == 0:
                f = 1
            else:
                f = c_freq[i-1] + 1
        
        c_freq.append(f)
    return c_freq
    
# print(c_freq)
```
