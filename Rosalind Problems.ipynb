# [Fibonacci](https://rosalind.info/problems/fib/)

```python
def wabbits(n, k):
    a, b = 1, 1
    for _ in range(2, n):
        a, b = b, b + k * a
    return b
```

```python
from Bio import SeqIO
sequences = [seq_record.seq for seq_record in SeqIO.parse("finding_shared_motif.fasta", "fasta")]
lengths = [len(seq) for seq in sequences]
print(sequences)
min_index = lengths.index(min(lengths))
reference_sequence = sequences[min_index]
sequences.pop(min_index)

print(reference_sequence)
```

```python
motifs = []

print()
for length in range(1, len(reference_sequence) + 1):
    for i in range(0, len(reference_sequence) - length + 1):
        candidate = reference_sequence[i:i+length]
        found_all = True
        for target_sequence in sequences:
            # print(f"{candidate=} {target_sequence=}")
            if target_sequence.find(str(candidate)) < 0:
                found_all = False
                break
        
        if found_all:
            # print(f"Appended: {candidate}")
            motifs.append(candidate)

print(motifs)
```

# [Introduction to Random Strings - Modeling Random Genomes](https://rosalind.info/problems/prob/)

```python
import math
DNA_String = (
    "GCCGGATCCCTTTGCCATTACCTGTCCAATGTTGTTTTACGTTCTCTCCCCCCGGTAGGAAGGACGAGGGAGGGTATTTCA"
)
A = [
    0.076,
    0.130,
    0.204,
    0.251,
    0.287,
    0.363,
    0.385,
    0.445,
    0.524,
    0.532,
    0.608,
    0.634,
    0.696,
    0.775,
    0.822,
    0.894,
    0.897
]
```

```python
B = []
for gc_content in A:
    
    g_p = c_p = gc_content / 2
    a_p = t_p = (1-gc_content) / 2
    
    prob_map = {
        "A": a_p,
        "C": c_p,
        "G": g_p,
        "T": t_p
    }
    
    probability = 1
    for neu in DNA_String:
        
        probability *= prob_map[neu]
        
    B.append(round(math.log10(probability), 3))
    
print(B)
```
